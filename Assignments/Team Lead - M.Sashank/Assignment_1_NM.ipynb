```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset

```python
data=pd.read_csv('House Price India.csv')
```

```python
data
```

# Univariate Analysis

```python
# Create a histogram of the price variable
plt.hist(data['Price'], bins=20)
plt.title('Histogram of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Univariate analysis of 'living area' variable using a boxplot
plt.boxplot(data['living area'])
plt.title('Living Area')
plt.show()
```

```python
# Univariate analysis of 'living area' variable using a density plot
plt.plot(data['living area'])
plt.title('Living Area')
plt.show()
```

```python
# Univariate analysis of 'waterfront present' variable using a bar chart
waterfront_counts = data['waterfront present'].value_counts()
plt.bar(waterfront_counts.index, waterfront_counts.values)
plt.title('Waterfront Present')
plt.show()
```

```python
# Univariate analysis of 'condition of the house' variable using a pie chart
condition_counts = data['condition of the house'].value_counts()
plt.pie(condition_counts.values, labels=condition_counts.index, autopct='%1.1f%%')
plt.title('Condition of the House')
plt.show()
```

```python
# Univariate analysis of 'living area' variable using a violin plot
plt.violinplot(data['living area'])
plt.title('Living Area')
plt.show()
```

# Bi-Variate Analysis

```python
# Bivariate analysis of 'living area' and 'price' variables using a scatter plot
plt.scatter(data['living area'], data['Price'])
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
# Bivariate analysis of 'built year' and 'price' variables using a line plot
grouped_data = data.groupby('Built Year')['Price'].mean()
plt.plot(grouped_data.index, grouped_data.values)
plt.xlabel('Built Year')
plt.ylabel('Price')
plt.show()
```

```python
# Bivariate analysis of 'waterfront present' and 'grade of the house' variables using a heatmap
grouped_data = data.groupby(['waterfront present', 'grade of the house'])['Price'].mean().unstack()
sns.heatmap(grouped_data, cmap='YlGnBu')
plt.xlabel('Grade of the House')
plt.ylabel('Waterfront Present')
plt.show()
```

# Multi-Variate Analysis

```python
# Multivariate analysis of 'living area', 'number of bedrooms', 'number of bathrooms', and 'price' variables using a scatter plot matrix
sns.pairplot(data[['living area', 'number of bedrooms', 'number of bathrooms', 'Price']])
plt.show()
```

```python
# Multivariate analysis of 'living area', 'number of bedrooms', 'number of bathrooms', and 'price' variables using a heatmap
correlation_matrix = data[['living area', 'number of bedrooms', 'number of bathrooms', 'Price']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')
plt.show()
```

```python
# Multivariate analysis of 'condition of the house', 'number of bedrooms', and 'price' variables using a bar chart
grouped_data = data.groupby(['condition of the house', 'number of bedrooms'])['Price'].mean().unstack()
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Condition of the House')
plt.ylabel('Price')
plt.show()
```
