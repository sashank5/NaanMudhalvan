```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
data=pd.read_csv('/content/House Price India.csv')
data
```

```python
# Choose the column to analyze
column_to_analyze = 'Price'
```

```python
# Print some basic statistics
print('Basic statistics for column', column_to_analyze)
print('-----------------------------------------')
print(data[column_to_analyze].describe())
```

Univariate Analysis

```python
# Plot a histogram
plt.hist(data[column_to_analyze], bins=20)
plt.title('Histogram of ' + column_to_analyze)
plt.xlabel(column_to_analyze)
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot a box plot
sns.boxplot(x=data[column_to_analyze])
plt.title('Box plot of ' + column_to_analyze)
plt.xlabel(column_to_analyze)
plt.show()
```

```python
# Plot a density plot
sns.kdeplot(data[column_to_analyze])
plt.title('Density plot of ' + column_to_analyze)
plt.xlabel(column_to_analyze)
plt.show()
```

Bi-Variate Analysis

```python
# Choose the columns to analyze
column_x = 'living area'
column_y = 'Price'
```

```python
# Print some basic statistics
print('Basic statistics for column', column_x)
print('-----------------------------------------')
print(data[column_x].describe())
```

```python
# Create a scatter plot
plt.scatter(data[column_x], data[column_y])
plt.title('Scatter plot of ' + column_x + ' vs. ' + column_y)
plt.xlabel(column_x)
plt.ylabel(column_y)
plt.show()
```

```python
# Create a box plot
sns.boxplot(x=data[column_x], y=data[column_y])
plt.title('Box plot of ' + column_x + ' vs. ' + column_y)
plt.xlabel(column_x)
plt.ylabel(column_y)
plt.show()
```

```python
# Create a violin plot
sns.violinplot(x=data[column_x], y=data[column_y])
plt.title('Violin plot of ' + column_x + ' vs. ' + column_y)
plt.xlabel(column_x)
plt.ylabel(column_y)
plt.show()
```

```python
# Calculate the correlation coefficient
corr = data[column_x].corr(data[column_y])
print('Correlation coefficient between', column_x, 'and', column_y, 'is', corr)
```

Multi-Variate Analysis

```python
# Choose the columns to analyze
columns_to_analyze = ['number of bedrooms', 'number of bathrooms', 'living area', 'Price']
```

```python
# Print some basic statistics
print('Basic statistics for columns')
print('-----------------------------------------')
print(data[columns_to_analyze].describe())
```

```python
# Create a pair plot
sns.pairplot(data[columns_to_analyze])
plt.show()
```

```python
# Create a heatmap
corr = data[columns_to_analyze].corr()
sns.heatmap(corr, annot=True)
plt.title('Heatmap of correlation matrix')
plt.show()
```
