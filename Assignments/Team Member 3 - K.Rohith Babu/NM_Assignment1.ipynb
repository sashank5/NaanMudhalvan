```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/House Price India.csv')
```

```python
# calculate basic descriptive statistics
mean = df.mean()
median = df.median()
std = df.std()
```

```python
df
```

performing univariate analysis

```python
# plot a histogram of the data
plt.hist(df, bins=10)
plt.title("Histogram of Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# plot a box plot of the data
plt.boxplot(df)
plt.title("Box Plot of Data")
plt.ylabel("Values")
plt.show()
```

```python
# plot a density plot of the data
df.plot(kind="density")
plt.title("Density Plot of Data")
plt.xlabel("Values")
plt.show()
```

performing bivariate analysis

```python
# extract x and y variables from the data
x = df["Price"]
y = df["lot area"]
```

```python
# calculate the correlation coefficient
corr_coef = x.corr(y)
```

```python
# plot a scatter plot of x and y
plt.scatter(x, y)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# plot a heatmap of the correlation matrix
corr_matrix = df.corr()
plt.imshow(corr_matrix, cmap='hot', interpolation='nearest')
plt.title("Heatmap of Correlation Matrix")
plt.colorbar()
plt.show()
```

performing Multivariate analysis

```python
# Choose the columns to analyze
columns_to_analyze = ['living area', 'lot area', 'number of floors', 'Price']
```

```python
# Create a pair plot
sns.pairplot(df[columns_to_analyze])
plt.show()
```

```python
# Create a heatmap
corr = df[columns_to_analyze].corr()
sns.heatmap(corr, annot=True)
plt.title('Heatmap of correlation matrix')
plt.show()
```

Performming descriptive statistics on the dataset

```python
# Compute descriptive statistics
mean = df.mean()
median = df.median()
mode = df.mode()
range = df.max() - df.min()
variance = df.var()
std_dev = df.std()
quartiles = df.quantile([0.25, 0.5, 0.75])
```

```python
# Print the results
print("Mean:\n", mean)
print("Median:\n", median)
print("Mode:\n", mode)
print("Range:\n", range)
print("Variance:\n", variance)
print("Standard Deviation:\n", std_dev)
print("Quartiles:\n", quartiles)
```
