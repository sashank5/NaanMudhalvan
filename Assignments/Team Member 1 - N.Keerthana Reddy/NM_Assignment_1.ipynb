```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/House Price India.csv')
```

Loading the dataset

```python
data
```

performing univariate analysis

```python
# Choose a variable to analyze
var_name = 'Price'
```

```python
# Create a histogram of the variable
plt.hist(data[var_name], bins=20)
plt.title(var_name + ' distribution')
plt.xlabel(var_name)
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a boxplot of the variable
sns.boxplot(x=data[var_name])
plt.title(var_name + ' distribution')
plt.xlabel(var_name)
plt.show()
```

```python
# Create summary statistics for the variable
print(data[var_name].describe())
```

performing Bi - Variate Analysis

```python
# extract x and y variables from the data
x = data["Price"]
y = data["Built Year"]
```

```python

# calculate the correlation coefficient
corr_coef = x.corr(y)
```

```python
# plot a scatter plot of x and y
plt.scatter(x, y)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# plot a heatmap of the correlation matrix
corr_matrix = data.corr()
plt.imshow(corr_matrix, cmap='hot', interpolation='nearest')
plt.title("Heatmap of Correlation Matrix")
plt.colorbar()
plt.show()
```

Performing Multivariate Analysis

```python
# Choose the columns to analyze
columns_to_analyze = ['living area', 'Built Year', 'number of floors', 'Price']
```

```python
# Create a pair plot
sns.pairplot(data[columns_to_analyze])
plt.show()
```

```python
# Create a heatmap
corr = data[columns_to_analyze].corr()
sns.heatmap(corr, annot=True)
plt.title('Heatmap of correlation matrix')
plt.show()
```

Performming descriptive statistics on the dataset

```python
# Calculate descriptive statistics for the 'Price' variable
price_stats = data['Price'].describe()
print('Descriptive Statistics for Price:')
print(price_stats)
```

```python
# Compute descriptive statistics
mean = data.mean()
median = data.median()
mode = data.mode()
range = data.max() - data.min()
variance = data.var()
std_dev = data.std()
quartiles = data.quantile([0.25, 0.5, 0.75])
```

```python
# Print the results
print("Mean:\n", mean)
print("Median:\n", median)
print("Mode:\n", mode)
print("Range:\n", range)
print("Variance:\n", variance)
print("Standard Deviation:\n", std_dev)
print("Quartiles:\n", quartiles)
```
